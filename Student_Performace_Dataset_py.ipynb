```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_excel("Students Performance Dataset.xlsx")

df.head()
```

```python
def performance_category(score):
    if score >= 75:
        return "Good"
    elif score >= 50:
        return "Average"
    else:
        return "Poor"

df["Performance"] = df["Total_Score"].apply(performance_category)
df.head()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x=df["Performance"], palette="viridis")
plt.title("Student Performance Distribution")
plt.xlabel("Performance Category")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x=df.index, y=df["Total_Score"], hue=df["Performance"], s=100)
plt.axhline(80, color='green', linestyle='--')
plt.axhline(60, color='orange', linestyle='--')
plt.title("Student Score Classification")
plt.xlabel("Student Number")
plt.ylabel("Total Score")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Total_Score"], bins=10)
plt.title("Distribution of Total Scores")
plt.xlabel("Total Score")
plt.ylabel("Count")
plt.show()
```
